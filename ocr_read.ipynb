```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import os
```

```python
#OpenCV is capable of performing advacned image processing like color change,creating an image array, blur, changing threshold of images etc.   
```

```python
# pip install pillow 
# pip install pytessaract 
import pytesseract 
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files (x86)\Tesseract-OCR\tesseract.exe'
```

```python
dir = r'C:\Users\somvi\Desktop\OCR reading\Train Images'
```

```python
img_list = []
for img in os.listdir(dir):
    image = cv2.imread(os.path.join(dir,img))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # by default cv2 uses BGR foramt and pytessaract assumes RGB so we convert the colors
    img_list.append(image)
```

```python
plt.imshow(img_list[0])
```

```python
img_text = []
for i in range(len(img_list)):
    ocr = pytesseract.image_to_string(img_list[i])
    img_text.append(ocr)
    
```
